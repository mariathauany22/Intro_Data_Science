### Atividade Complementar


Imagine que você é o responsável pelo controle de estoque e pretende criar um sistema que registre os pedidos das pessoas ou informe que não há estoque suficiente para atender a demanda.

Nesse caso digamos que os produtos que estão disponíveis no estoque sejam Sabão (com 75 unidades disponíveis); Detergente (com 30 unidades disponíveis) e Desinfetante (com 50 unidades disponíveis).

Crie um código em que seja pedido ao usuário que informe o produto e a quantidade desejada.

Ao final o sistema precisa analisar o pedido com os valores disponíveis no estoque e informar se o pedido foi feito com sucesso ou não foi possível atender o pedido devido à quantidade em estoque ser inferior à demandada.

```python
estoque = {
    "sabao": 75,
    "detergente": 30,
    "desinfetante": 50
}

produto = input("Qual produto deseja (sabao, detergente, desinfetante)? ").lower()
quantidade = int(input("Informe a quantidade desejada: "))

if produto in estoque:
    if estoque[produto] >= quantidade:
        print(f"Pedido de {quantidade} unidade(s) de {produto} feito com sucesso!")
        estoque[produto] -= quantidade
    else:
        print(f"Não foi possível atender o pedido. Quantidade em estoque ({estoque[produto]}) insuficiente.")
else:
    print("Produto não cadastrado no estoque.")
```

#### 6 - Crie linhas de código em que o usuário deve informar apenas os números do CPF (que contém 11 caracteres). Nessas linhas de código você precisa criar um alerta de erro caso o usuário preencha os caracteres "." ou "-". Depois disso crie um código que mostre na tela o CPF no formato tradicional (xxx.xxx.xxx-xx). Resumindo: O usuário precisa informar apenas os números do CPF(xxxxxxxxxxx), mas, no momento da exibição na tela o CPF informado deve estar no formato xxx.xxx.xxx-xx.

```python
cpf = input("Informe os 11 números do CPF (apenas números): ")

if "." in cpf or "-" in cpf:
    print("ERRO: Favor preencher apenas números, sem '.' ou '-'.")
elif len(cpf) != 11 or not cpf.isdigit():
    print("ERRO: O CPF deve conter exatamente 11 números.")
else:
    cpf_formatado = f"{cpf[:3]}.{cpf[3:6]}.{cpf[6:9]}-{cpf[9:]}"
    print(f"CPF: {cpf_formatado}")
```

```python
produtos = ["caneta", "mouse", "lapis", "borracha", "grampeador", "cartucho"]

print(f"1 - Há {len(produtos)} produtos diferentes no estoque.")
```

### Digamos que o setor onde você trabalha quer instituir um controle de almoxarifado e possui as seguintes informações: produtos = ["canettta", "mouseeeeee", "lapis", "borracha", "grampeador", "cartuxxo"] quantidades = [250, 50, 320, 450, 70, 190] Com base nessas informações responda as questões 5 a 10:

5 - Crie um código que faça a correção dos erros de digitação ocorridos na lista "produtos".

```python
produtos = ["canettta", "mouseeeeee", "lapis", "borracha", "grampeador", "cartuxxo"]
quantidades = [250, 50, 320, 450, 70, 190]

produtos[0] = "caneta"
produtos[1] = "mouse"
produtos[5] = "cartucho"
print(f"5 - Lista corrigida: {produtos}")
```

6 - Crie um sistema onde seja possível um usuário consultar quantas unidades de um produto constam no almoxarifado. Nesse sistema o usuário informa o nome do produto e o sistema deve informar o seguinte:

Há XX unidades do produto YY no almoxarifado.

importante: considere que o usuário sempre informará um produto que consta no estoque.

```python
busca = input("6 - Qual produto deseja consultar? (Existe na base): ")
qtd = quantidades[produtos.index(busca)]
print(f"Há {qtd} unidades do produto {busca} no almoxarifado.")
```

7 - Semelhante à questão 6, crie um sistema onde seja possível um usuário informar o produto que ele deseja consultar no almoxarifado. Nesse sistema o usuário informa o nome do produto e o sistema deve se comportar das seguintes formas:

Quando o produto informado existir no almoxarifado, informar: "Há XX unidades do produto YY no almoxarifado."

Quando o produto informado não existir no almoxarifado ou o usuário informar um produto com erro de digitação: "O produto ZZ não consta no almoxarifado."


```python
busca_real = input("7 - Qual produto deseja consultar? ")
if busca_real in produtos:
    qtd_real = quantidades[produtos.index(busca_real)]
    print(f"Há {qtd_real} unidades do produto {busca_real} no almoxarifado.")
else:
    print(f"O produto {busca_real} não consta no almoxarifado.")
```

8 - Crie linhas de código que permitam identificar qual o produto que possui mais unidades e qual possui menos unidades no almoxarifado.

```python
max_qtd = max(quantidades)
min_qtd = min(quantidades)
prod_mais = produtos[quantidades.index(max_qtd)]
prod_menos = produtos[quantidades.index(min_qtd)]
print(f"8 - Produto com mais unidades: {prod_mais} ({max_qtd}). Produto com menos: {prod_menos} ({min_qtd}).")
```

9 - Crie um código que permita saber a quantidade total de produtos que há no almoxarifado, ou seja, o sistema deve somar todos os itens da lista "quantidades".

10 - Digamos que o setor não utilizará mais o produto "cartucho" e irá trocar com outro setor todos esses "cartuchos" por 30 unidades de "teclado" para computadores. ### EXERCÍCIO COMPLEMENTAR

Digamos que você trabalha num setor que atende o público por meio de processos administrativos. As listas representam os meses e as quantidades de processos atendidos, respectivamente:

meses = ["janeiro", "fevereiro", "março", "abril", "maio", "junho", "julho", "agosto", "setembro", "outubro", "novembro", "dezembro"]

atendimentos_semestre_1 = [350, 280, 420, 560, 325, 295]
atendimentos_semestre_2 = [490, 550, 620, 516, 380, 478]

Crie um sistema onde seja possível evidenciar na tela o meses em que ocorreram as maiores e menores quantidades de atendimentos, bem como evidenciar o percentual que esses atendimentos representam em relação ao total de atendimentos no ano.

No sistema deve ser evidenciada as seguintes menagens:


"O mês de xxx foi o que registrou a maior quantidade de atendimentos no ano com yyy atendimentos, o que representa zz % do total de atendimentos no ano."

"O mês de xxx foi o que registrou a menor quantidade de atendimentos no ano com yyy atendimentos, o que representa apenas zz % do total de atendimentos no ano."

```python
meses = ["janeiro", "fevereiro", "março", "abril", "maio", "junho",
         "julho", "agosto", "setembro", "outubro", "novembro", "dezembro"]

atendimentos_semestre_1 = [350, 280, 420, 560, 325, 295]
atendimentos_semestre_2 = [490, 550, 620, 516, 380, 478]
atendimentos_ano = atendimentos_semestre_1 + atendimentos_semestre_2

total_atendimentos = sum(atendimentos_ano)
maior_qtd = max(atendimentos_ano)
menor_qtd = min(atendimentos_ano)

mes_maior = meses[atendimentos_ano.index(maior_qtd)]
mes_menor = meses[atendimentos_ano.index(menor_qtd)]

perc_maior = (maior_qtd / total_atendimentos) * 100
perc_menor = (menor_qtd / total_atendimentos) * 100

print(f"O mês de {mes_maior} foi o que registrou a maior quantidade de atendimentos no ano com {maior_qtd} atendimentos, o que representa {perc_maior:.2f}% do total de atendimentos no ano.")

print(f"O mês de {mes_menor} foi o que registrou a menor quantidade de atendimentos no ano com {menor_qtd} atendimentos, o que representa apenas {perc_menor:.2f}% do total de atendimentos no ano.")
```
